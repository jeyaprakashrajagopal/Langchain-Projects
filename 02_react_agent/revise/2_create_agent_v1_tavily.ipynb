```python
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from langchain_tavily import TavilySearch
from dotenv import load_dotenv
from langchain.agents import create_agent
from langsmith import Client
from langchain_core.messages import HumanMessage
from langchain.tools import tool
from tavily import TavilyClient

load_dotenv()

tavily = TavilyClient()
```

```python
@tool
def search(query: str) -> str:
    """Tool that searches the query over the internet 
    Args:
        query: The actual search query
    Returns:
        str: returns a string output
    """
    return tavily.search(query=query)
```

```python
def main():
    print("Search with tavily!")
    llm = ChatOpenAI()
    tools = [search]

    agent = create_agent(model=llm, tools=tools)

    result = agent.invoke({"messages": HumanMessage("What is the weather in frankfurt?")})
    print(result)
```

```python
if __name__ == "__main__":
    main()
```
