```python
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from langchain_tavily import TavilySearch
from dotenv import load_dotenv
from langchain.agents import create_agent
from langsmith import Client
from langchain_core.messages import HumanMessage
from langchain.tools import tool
from tavily import TavilyClient

load_dotenv()

tavily = TavilyClient()
```

```python
@tool
def search(query: str) -> str:
    """Tool that searches the query over the internet 
    Args:
        query: The actual search query
    Returns:
        str: returns a string output
    """
    return tavily.search(query=query)
```

```python
from typing import List
from pydantic import Field, BaseModel

class Source(BaseModel):
    """Schema for a source that the agent uses"""
    url: str = Field(description="The URL of the source used by an agent")

class AgentResponse(BaseModel):
    """Schema for agent's response to the query"""
    answer: str = Field(description="Agent's answer to the given query")
    sources: List[Source] = Field(default_factory=list, description="list of sources used by the agent")
```

```python
def main():
    print("Search with tavily!")
    llm = ChatOpenAI()
    tools = [search]

    agent = create_agent(model=llm, tools=tools, response_format=AgentResponse)

    result = agent.invoke({"messages": HumanMessage("Search 3 AI engineer jobs in frankfurt area in linkedin")})
    print(result)

if __name__ == "__main__":
    main()
```
