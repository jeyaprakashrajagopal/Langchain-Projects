```python
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from langchain_tavily import TavilySearch
from dotenv import load_dotenv
from langchain.agents import create_agent
from langsmith import Client
from langchain_core.messages import HumanMessage
from langchain.tools import tool

load_dotenv()
```

```python

@tool
def search(query: str) -> str:
    """Tool that searches query in the internet 
    
    Args:
        query: the query to search for
    Returns:
        returns the search result as string
    """
    return "Frankfurt has a low temperature and has a warning for skidding!"

def main():
    print("Hello from create agent latest version!")
    tools = [search]
    llm = ChatOpenAI()
    agent = create_agent(model=llm, tools=tools)
    result = agent.invoke({"messages": HumanMessage("what is the weather in Frankfurt?")})
    print(result)


if __name__ == '__main__':
    main()
```
